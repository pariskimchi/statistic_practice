###  가설검정(hypothesis Testing)  
=> 귀무가설(Null hypothesis: HO) and 대립가설(Alternative Hypothesis:H1)

    1. 귀무가설
    ==> 모집단의 특성에 대해 옳다고 제안하는 잠정적인 주장 
    
    2 . 대립가설 
    ==> 귀무가설이 거짓이라면 ==> 대립가설은 참이 되는것 
    
#### 즉 귀무가설과 대립가설 둘중에 뭘 선택해야되는지 둘때, 
    => 귀무가설이 맞으면 귀무가설, 귀무가설이 틀리면 대립가설을 선택 
    => 이 선택 기준은 p value(유의 확률)을 기준으로 한다

### 가정 1: 
    1. 새로운 페이지 vs 기존 페이지

    H1(대립가설) : 새로운 페이지 > 기존페이지

## Type of errors 
    1. Type 1 error(1종 오류)?
    => 귀무가설(HO)가 실제로는 참이어서 채택해야 함에도 불구하고 
    표본의 오차때문에 이를 채택하지 않는 오류=> 이걸 유의 수준이라고 부른다 
    => a
    
    2. Type 2 error(2종 오류) 
    => 귀무가설(H0)가 거짓이라서 채택하지 말아야 하는데 
    표본의 오차떄문에 이를 채택하는 오류 
    ==> b (베타)

### 1종 오류
    => 대립가설이 선택된것,(귀무가설이 옳음에도 불구하고)
    => False Positive

## 2종 오류 
    => False negative)
