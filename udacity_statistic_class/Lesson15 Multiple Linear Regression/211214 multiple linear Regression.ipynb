## Fitting Multiple Linear regression Model 

```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm; 

df = pd.read_csv('./house_prices.csv')
df.head()
```

## 1. Using statsmodels, fit 3 simple linear regression models to predict price. 
    => area, bedrooms, bathroomes

```python
df['intercept'] = 1

lm = sm.OLS(df['price'],df[['intercept','area']])
results = lm.fit()
results.summary()
```

```python
# bedrooms 

lm = sm.OLS(df['price'],df[['intercept','bedrooms']])
results = lm.fit()
results.summary()
```

```python
lm = sm.OLS(df['price'],df[['intercept','bathrooms']])
results = lm.fit()
results.summary()
```

## 2. make a  multiplie linear regression models

```python
lm = sm.OLS(df['price'],df[['intercept','area','bedrooms','bathrooms']])
results = lm.fit()
results.summary()
```

## 3. adding style 

```python
lm = sm.OLS(df['price'],df[['intercept','area','bedrooms','bathrooms','style']])
results = lm.fit()
results.summary()
```
