## interpreting Coefficients 


```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm;

df = pd.read_csv('./house_prices.csv')
df.head()
```

### Model 1. 
    => predict price using neighborhood, style, and area 

```python
df['neighborhood'].unique(), df['style'].unique()
```

```python
dummies1_df = pd.get_dummies(df['neighborhood'])
dummies2_df = pd.get_dummies(df['style'])

new_df = df.join(dummies1_df)
new_df = new_df.join(dummies2_df)
new_df.head()
```

```python
new_df['intercept'] = 1

lm = sm.OLS(new_df['price'],new_df[['intercept','area','A','B','ranch','victorian']])
results = lm.fit()
results.summary()
```

## model 2 

    => predict price 
    =>, using area , area squared
    => using style, 
    => not neighborhood 
    

```python
new_df['area_squared'] = new_df['area']*new_df['area']
```

```python
new_df.head()
```

```python
lm = sm.OLS(new_df['price'],new_df[['intercept','ranch','victorian','area','area_squared']])
results = lm.fit()
results.summary()
```
